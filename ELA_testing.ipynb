Importo la clase ELA

```python
from ELA import ELA
```

Instancio la clase ELA

```python
ela_obj = ELA('https://cloudfront-eu-central-1.images.arcpublishing.com/prisa/J32UN3N5ENDBBBC7IHYKCT2JOQ.jpg')

probability, num_of_regions, labeled_regions = ela_obj.calculate_estimated_edit_percentaje()

print("Porcentaje estimado de edición de la imagen: " + str(probability))
print("Numero estimado de regiones totales " + str(num_of_regions))
```

Ahora pruebo las funciones que me muestren las imagenes

```python
ela_obj.show_original_image()
```

```python
ela_obj.show_ela_image()
```
