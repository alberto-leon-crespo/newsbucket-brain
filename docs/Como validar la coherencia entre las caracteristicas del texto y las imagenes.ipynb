Como obtener la puntuación de similitud entre los metodos de NLP y de analisis de imagenenes

```python
from sklearn.metrics.pairwise import cosine_similarity

# Extracted features from image and text (already defined)
image_features = extract_image_features(image_path)
text_features = extract_text_features(news_text)

# Calculate cosine similarity between image and text features
similarity_score = cosine_similarity([image_features], [text_features])[0][0]

# Define the coherence threshold
coherence_threshold = 0.7  # Example threshold, adjust according to your needs

# Make a decision based on the similarity score and coherence threshold
if similarity_score >= coherence_threshold:
    print("Image and text are coherent.")
else:
    print("Image and text are incoherent.")
```
